## Importation des librairies

```python
import cv2
import face_recognition
import matplotlib.pyplot as plt
```

## Lecture d'une image

### Avec cv2

```python
sadio = cv2.imread("sadio.jpg")
plt.imshow(sadio)
```

```python
sadio = cv2.imread("sadio.jpg")
sadio = cv2.cvtColor(sadio, cv2.COLOR_BGR2RGB)
plt.imshow(sadio)
```

### Avec face_recognition

```python
sadio1 = face_recognition.load_image_file("sadio.jpg")
plt.imshow(sadio1)
```

```python
list_faces = face_recognition.face_locations(sadio1)
list_faces
```

```python
coord_sadio = list_faces[0]
coord_sadio
```

```python
top, right, bottom, left = coord_sadio[0], coord_sadio[1], coord_sadio[2], coord_sadio[3]
```

```python
sadio_rect = cv2.rectangle(sadio1, (left, top), (right, bottom), (255,0,0), 4)
```

```python
plt.imshow(sadio_rect)
```

```python
ronaldo = face_recognition.load_image_file("Ronaldo.jpg")
plt.imshow(ronaldo)
```

### Comparaison entre deux images

#### Encodage des images

```python
sadio_encod = face_recognition.face_encodings(sadio1)[0]
ronaldo_encod = face_recognition.face_encodings(ronaldo)[0]
```

#### Comparaison

##### Comparaison Sadio - Ronaldo

```python
result = face_recognition.compare_faces([sadio_encod], ronaldo_encod)
```

```python
result
```

##### Comparaison Sadio - Sadio

```python
sadio2 = face_recognition.load_image_file("sadio_2.jpg")
plt.imshow(sadio2)
```

```python
sadio2_encod = face_recognition.face_encodings(sadio2)[0]
```

```python
face_recognition.compare_faces([sadio_encod], sadio2_encod)
```

#### Ecriture sur une photo

```python
sadio_avec_nom = cv2.putText(sadio_rect, "Sadio", (left , top-10),cv2.FONT_HERSHEY_PLAIN, 6, (0,255,0))
```

```python
plt.imshow(sadio_avec_nom)
```
